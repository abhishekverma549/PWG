```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_excel('Data.xlsx', na_values = ['..'], usecols=range(0,23), )
df = df.rename(columns={'2000 [YR2000]':'2000','2001 [YR2001]':'2001','2002 [YR2002]':'2002','2003 [YR2003]':'2003',
                        '2004 [YR2004]':'2004','2005 [YR2005]':'2005','2006 [YR2006]':'2006','2007 [YR2007]':'2007',
                        '2008 [YR2008]':'2008','2009 [YR2009]':'2009','2010 [YR2010]':'2010','2011 [YR2011]':'2011',
                        '2012 [YR2012]':'2012','2013 [YR2013]':'2013','2014 [YR2014]':'2014','2015 [YR2015]':'2015',
                        '2016 [YR2016]':'2016','2017 [YR2017]':'2017','2018 [YR2018]':'2018',})
df.dropna(how = 'all', inplace  =True)  
```

```python
Country = []
New_Index = df['Country Name']
for item in New_Index:
    Country.append(item.upper())  
df['Country'] = Country
```

```python
df.drop(['Country Name'], axis = 1, inplace=True)
```

```python
df.set_index('Country', inplace = True)
```

```python
X_Axis = df.columns[3:]
X_indexes = np.arange(len(X_Axis))
```

# Access to electricity in Rural/ Urban population

```python
Country_Name = input('Enter Country name: ')
Country_Name = Country_Name.upper()

try:
    def CHN_JPN(Index_Name):
        
        Country = df.loc[Index_Name]
        
        plt.figure(figsize=(10, 6), dpi=100, facecolor='w')
        width = 0.25
        
        plt.style.use('seaborn-darkgrid')
        plt.bar(X_indexes-width, Country.iloc[0][3:], width = width, label = 'Access to electricity in Urban population')
        plt.bar(X_indexes+width, Country.iloc[1][3:], width = width, label = 'Access to electricity in Rural population')
        
        plt.xticks(ticks = X_indexes, labels = X_Axis)
        plt.xlabel('Years')
        plt.ylabel('Access to electricity in Urban/ Rural population')
        plt.title("{}'s Rural vs Urban population graph who have access to electricty".format(Country_Name))
        plt.grid(True)
        plt.tight_layout()
        #plt.savefig('{}.png'.format(C_Name))
        

    CHN_JPN(Country_Name)
    
except:
    print('\nOops Entered country not available!')
```
